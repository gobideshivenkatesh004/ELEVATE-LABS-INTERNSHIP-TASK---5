```python
# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 2. Load CSV using Pandas
try:
    df = pd.read_csv('/content/sales_data_sample (1).csv', encoding='latin1')
except UnicodeDecodeError:
    df = pd.read_csv('/content/sales_data_sample (1).csv', encoding='cp1252')
```

```python
# 3. Total sales by product line
product_sales = df.groupby('PRODUCTLINE')['SALES'].sum()
print("\nTotal Sales by Product Line:")
print(product_sales)
```

```python
# 4. Total sales by product and region
product_region_sales = df.groupby(['PRODUCTLINE', 'COUNTRY'])['SALES'].sum()
print("\nTotal Sales by Product and Region:")
print(product_region_sales)
```

```python
# 5. Plot total sales by product
product_sales.plot(kind='bar', title='Total Sales by Product', figsize=(10, 6), color='skyblue')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
# 6. Optional: Plot sales by product and region
product_region_sales.unstack().plot(kind='bar', figsize=(10, 6))
plt.title('Sales by Product and Region')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
